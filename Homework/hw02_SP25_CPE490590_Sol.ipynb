# 1. Linear Regression and Regularization (15 Points)
## Solution:

$$
\argmin_\textbf{w} \bigg\{ ||\textbf{y} - \textbf{x} \textbf{w}||^2  + \lambda||\textbf{w}||^2  \bigg\}

$$

$$

\cfrac{d}{d\textbf{w}}\Bigg\{   \cfrac{1}{2}(\textbf{y} - \textbf{x}\textbf{w})^\top(\textbf{y} - \textbf{x}\textbf{w})  + \cfrac{\lambda}{2}\textbf{w}^\top \textbf{w}\Bigg\} = - \textbf{x}^\top (\textbf{y} -\textbf{x}\textbf{w}) + \lambda \textbf{w} = 0
$$

$$
- \textbf{x}^\top (\textbf{y} -\textbf{x}\textbf{w}) + \lambda \textbf{I}\textbf{w} = 0\\
(\textbf{x}^\top\textbf{x} + \lambda\textbf{I})\textbf{w} = \textbf{x}^\top \textbf{y}\\
\textbf{w} = (\textbf{x}^\top\textbf{x} + \lambda \textbf{I})^{-1} \textbf{x}^\top \textbf{y}
$$

The original solution (without regularization) is $\textbf{w} = (\textbf{x}^\top\textbf{x})^{-1} \textbf{x}^\top \textbf{y}$. Compared to this, with regularization, an extra term is $ \lambda \textbf{I}$. The impact of adding penalty is that it penalizes any data points far from the regressed line, so that the slope of the line is not affected too much, and tends to be less steep. This results in model preventing from overfitting.

# 2. Optimizing a Function using Partial Differentiation (5 points)

## Solution:

\begin{equation}
    J(w_1) = (w_1 -12)^2 + 32
\end{equation}


To find the value of $w_1$ that minimizes the cost function $J(w_1)$, we can take the derivative of $J(w_1)$ with respect to $w_1$, set it to zero, and solve for $w_1$. This is because the minimum or maximum of a function occurs where its derivative is zero.

The derivative of $J(w_1)$ with respect to $w_1$ is:

$$
\frac{dJ(w_1)}{dw_1} = 2(w_1 - 12)
$$

Setting this to zero gives:

$$
2(w_1 - 12) = 0
$$

Solving for $w_1$ gives:

$$
w_1 = 12
$$

So, the value of $w_1$ that minimizes the cost function $J(w_1)$ is $w_1 = 12$.

# 4. Goodness of Fit (5 Points)
What is goodness of fit in the linear regression and how do we quantify them?

## Solution:

 The goodness of fit refers to how well the regression model fits the observed data. 
 Coefficient of Determination ($R^2$)  is one of the most common metrics for evaluating the goodness of fit. It ranges from 0 to 1 and represents the proportion of variance in the dependent variable that can be predicted from the independent variable(s). An $R^2$ of 1 indicates that the regression predictions perfectly fit the data. $R^2$ is given by

  \begin{equation}
    R^2 = 1-\cfrac{\textrm{SSE}}{\textrm{SSTO}}
\end{equation}

where $\textrm{SSTO (Total Sum Square)} = \sum (y_i - \bar{y})^2$, and $\textrm{SSE (Sum of Squared Errors)} = \sum (y_i -\hat{y})^2$.


# 5. Constraint Optimization (10 points)

Consider the function $f(x, y) = e^{xy}$ subject to constraint $x^3 + y^3 = 16$. Use Lagrange multipliers to find the coordinates $(x, y)$ of any points subject to the given constraint where the function $f$ could attain a maximum or minimum.

## Solution:

We wish to find the extreme values of the function $f$ subject to the constraint $g(x,y) = x^3 + y^3 - 16$

Using Lagrange multiplier:

\begin{equation}
    \begin{aligned}
        \nabla f & = \lambda \nabla g \\
        \Rightarrow ye^{xy} & = 3\lambda x^2\\
        \Rightarrow xe^{xy} & = 3\lambda y^2\\
      \end{aligned}
\end{equation}

In this case, if with either $x=0$ or $y=0$, it would make $x=0, y = 0$ based on Equation~\eqref{eq:eq1} and would violate our constraint $g(x,y) = x^3 + y^3 - 16$. Hence, clearly $x \neq 0$, $y \neq 0$.

Then

$$
\lambda = \cfrac{ye^{xy}}{3x^2} = \cfrac{xe^{xy}}{3y^2}
$$
based on Equation~\eqref{eq:eq1}.

That gives as $x^3 = y^3 \Rightarrow x = y$. Putting back this in our constraint, we get $2x^3 = 16 \Rightarrow x = 2, \Rightarrow y  = 2$.

Thus, at  the point $(2, 2)$, a point can attain the optimum value.

```python
import numpy as np
# Define the objective function
def objective(x):
    x, y = x
    return torch.exp(x*y)

# Define the constraint function
def constraint(x):
    x, y = x
    return x**3 + y**3 - 16

# Penalty method: Combine objective and constraint
def loss_fn(x, penalty_weight=10.0):
    obj = objective(x)
    constr_violation = constraint(x)**2  # Squared to ensure non-negativity
    return obj + penalty_weight * constr_violation

# Initialize trainable parameters
x = torch.tensor([1.0, 1.0], requires_grad=True)  # Initial guess for x and y

# Set up the optimizer
optimizer = torch.optim.SGD([x], lr=0.00001)  # Learning rate can be adjusted

# Perform gradient descent
for i in range(5000):
    # Zero the gradients
    optimizer.zero_grad()

    # Compute the loss
    loss = loss_fn(x)

    # Backpropagate to compute gradients
    loss.backward()

    # Update the parameters
    optimizer.step()

    # Print progress (optional)
    if (i + 1) % 500 == 0:
        print(f"Iteration {i + 1}: x = {x[0].item():.4f}, y = {x[1].item():.4f}, Loss = {loss.item():.4f}")

# Print the final results
print("\nOptimal values:")
print(f"x = {x[0].item():.4f}")
print(f"y = {x[1].item():.4f}")
print(f"Objective value = {objective(x).item():.4f}")
print(f"Constraint violation = {constraint(x).item():.4f}")

```

### Lagrange Multiplier Method:
To solve the optimization problem for the function \( f(x, y) = e^{xy} \) subject to the constraint \( x^3 + y^3 = 16 \) using the method of Lagrange multipliers, follow these steps:

The Lagrangian function is:

$$
\mathcal{L}(x, y, \lambda) = e^{xy} - \lambda (x^3 + y^3 - 16)
$$

Take the partial derivatives of \(\mathcal{L}\) with respect to \(x\), \(y\), and \(\lambda\), and set them equal to zero:

$$
\frac{\partial \mathcal{L}}{\partial x} = y e^{xy} - 3\lambda x^2 = 0
$$

$$
\frac{\partial \mathcal{L}}{\partial y} = x e^{xy} - 3\lambda y^2 = 0
$$

$$
\frac{\partial \mathcal{L}}{\partial \lambda} = -(x^3 + y^3 - 16) = 0
$$

From the first two equations, eliminate \(\lambda\) by dividing:

$$
\frac{y}{x} = \frac{x^2}{y^2} \implies y^3 = x^3
$$

Substitute \(y^3 = x^3\) into the constraint \(x^3 + y^3 = 16\):

$$
2x^3 = 16 \implies x^3 = 8 \implies x = 2
$$

Since \(y^3 = x^3\), we also have \(y = 2\).

The critical point is \((x, y) = (2, 2)\).

Substitute \(x = 2\) and \(y = 2\) into \(f(x, y)\):

$$
f(2, 2) = e^{2 \times 2} = e^4
$$

\end{equation}

where $x_1$ and $x_2$ are raidus and height of cylinder and $V$ is the required volume.

Choose a few values of $V$ as 30.0, 40.0, 50.0, and 60.0

and perform constraint minimization using Project Gradient method in PyTorch. 

Choose appropriate initial values of $x_1$ and $x_2$ and learning rate of $0.001$.

Is there any relationship between different values of $x_1$  that minimizes $f$ and $V$ that you observe?

## Solution:

```python
import torch
import numpy as np
import matplotlib.pyplot as plt


# Objective function
def f(x1, x2):
    return 2 * torch.pi * x1 * (x1 + x2)

# Constraint (used to compute x2 from x1)
def compute_x2(x1, V):
    return V / (torch.pi * x1**2)

# Projected Gradient Descent (only x1 is optimized)
def projected_gradient_descent(V, lr=0.001, max_iter=10000, tol=1e-6):
    # Initialize x1 (radius) with gradient tracking
    x1 = torch.tensor(2.0, requires_grad=True)
    
    for _ in range(max_iter):
        # Compute x2 (height) from x1 using the constraint
        x2 = compute_x2(x1, V)
        
        # Compute the surface area (loss)
        loss = f(x1, x2)
        
        # Compute gradients of loss w.r.t. x1
        loss.backward()
        
        # Gradient descent step for x1
        with torch.no_grad():
            x1 -= lr * x1.grad
            x1.grad.zero_()  # Clear gradients for next iteration
        
        # Check convergence (optional)
        if torch.abs(loss) < tol:
            break
    
    # Final x2 after optimization
    x2 = compute_x2(x1, V)
    return x1.item(), x2.item()

# Test for different values of V
V_values = [30.0, 40.0, 50.0, 60.0]
results = {}

x1vals = []
x2vals = []
for V in V_values:
    x1_opt, x2_opt = projected_gradient_descent(V)
    x1vals.append(x1_opt)
    x2vals.append(x2_opt)
    results[V] = (x1_opt, x2_opt)
    print(f"For V = {V}: x1 = {x1_opt:.4f}, x2 = {x2_opt:.4f}")

fig = plt.figure(figsize=(6, 7))
plt.plot(V_values, x1vals)
plt.xlabel('V')
plt.xlabel('$x_1$')


fig = plt.figure(figsize=(6, 7))
plt.plot(V_values, x2vals)
plt.xlabel('V')
plt.xlabel('$x_2$')


```

```python
import matplotlib.pyplot as plt
import torch
import pandas as pd
plt.rcParams['font.family'] = 'Serif'
plt.rcParams['font.size'] = 15

Toluca = pd.read_csv('https://github.com/rahulbhadani/CPE490_590_Sp2025/raw/refs/heads/master/Data/Toluca/toluca.txt', sep='\t')
Toluca
```

```python
# Convert data to PyTorch tensors
x = Toluca['LotSize'].to_numpy()
y = Toluca['WorkHours'].to_numpy()

fig, ax = plt.subplots(figsize=(7, 4))
plt.scatter(x, y, c='orange', edgecolors='blue', s=100)
fig.patch.set_alpha(0.0)
plt.xlabel('x: Lot Size')
plt.ylabel('y: Work Hours')

X = torch.tensor(x, dtype=torch.float32).view(-1, 1)
Y = torch.tensor(y, dtype=torch.float32).view(-1, 1)

# Initialize parameters (weights)
w_1 = torch.randn(1, requires_grad=True, dtype=torch.float32)  # Weight for X
w_0 = torch.randn(1, requires_grad=True, dtype=torch.float32)  # Bias term

# Lists to store w_1, w_0 values over epochs
w_1_vals = []
w_0_vals = []
losses = []

# Define the Linear Regression model manually
def model(X):
    return w_1 * X + w_0

# Loss function: Mean Squared Error
def mse_loss(Y_pred, Y_true):
    return torch.mean((Y_pred - Y_true)**2)


# Optimizer: Using SGD (Stochastic Gradient Descent)
learning_rate = 0.0001
epochs = 1000
for epoch in range(epochs):
    # Forward pass: Compute predicted y
    Y_pred = model(X)
    
    # Compute loss
    loss = mse_loss(Y_pred, Y)
    losses.append(loss.item())
    # Backward pass: Compute gradients
    loss.backward()
    
    # Update parameters manually
    with torch.no_grad():
        w_1 -= learning_rate * w_1.grad
        w_0 -= learning_rate * w_0.grad
        
        # Zero the gradients after updating
        w_1.grad.zero_()
        w_0.grad.zero_()
        
    # Store the values of w_1 and w_0 at each epoch
    w_1_vals.append(w_1.item())
    w_0_vals.append(w_0.item())
    
    # Print the loss every 10 epochs
    if (epoch+1) % 100 == 0:
        print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}')

# Print the learned parameters
print(f'Learned weight: {w_1.item():.4f}')
print(f'Learned bias: {w_0.item():.4f}')
```

# 14. Solving a Diï¬€erential Equation by FiÂ‚ing a Polynomial using PyTorch (20 Points)

Following differential equation models radioactive decay
\begin{equation}
  \cfrac{d N}{d t} + k N = 0
\end{equation}

where $N(t)$ is the number of radioactive atoms in some radioactive material.


Assume $k = 1.216 \times 10^{-2}$, 

Its analytical solution is given by 

$$
N(t) = N(0)e^{-kt}
$$

Compute $N(t)$ using PyTorch by fitting a Polynomial of degree $3$ to the above differential equation. Choose the range of $t$ as $t =0.0$ to $t = 200.0$ sampled at $t = 0.004$ seconds. Consider $N(0) = 5.0$.

Choose learning rate as $0.001$ and number of epochs as $500000$.

The half life $t_{1/2}$ is given by 
$t_{1/2} = \cfrac{\ln 2}{k}$ at which number of radioactive atoms become half.

Compute the number of atoms at half life from the fitted model and compare with the one you obtain from the analytical solution.


Will you be able to make a prediction on $t= 5000.0$ correctly? Why or why not?



```python
import torch
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define the polynomial parameterization of y(x)
class PolynomialModel:
    def __init__(self, degree):
        # Initialize coefficients as trainable parameters
        self.coeffs = torch.randn(degree + 1, requires_grad=True)

    def forward(self, t):
        # Compute N(t) = c_0 + c_1*t + c_2*t^2 + ...
        N = torch.zeros_like(t)
        for i, coeff in enumerate(self.coeffs):
            N += coeff * (t ** i)
        return N

# Step 2: Define the loss function (residual of the ODE + initial condition penalty)
def ode_loss(model, t, k, N0):
    N = model.forward(t)  # Predict N(t)
    dN_dt = torch.autograd.grad(N, t, grad_outputs=torch.ones_like(N), create_graph=True)[0]  # Compute dN/dt
    
    residual = dN_dt + k * N
    
    # Penalty for initial condition y(0) = 1
    initial_condition_penalty = (model.forward(torch.tensor([[0.0]], requires_grad=True)) - N0)**2
    
    # Total loss: mean squared residual + initial condition penalty
    return torch.mean(residual**2) + initial_condition_penalty

# Step 3: Training setup
degree = 3  # Degree of the polynomial
model = PolynomialModel(degree)


k = 1.216e-2  # Decay constant
N0 = 5.0      # Initial number of atoms
lr = 0.001    # Learning rate
num_epochs = 500000

optimizer = optim.Adam([model.coeffs], lr=lr)

# Generate training data (t values)
t_data = torch.arange(0, 200, 0.004)

#linspace(0, 200, 50000).view(-1, 1)  # t ranges from 0 to 2
t_data.requires_grad_(True)  # Ensure t_data requires gradients for autograd

# Step 4: Training loop
for epoch in range(num_epochs):
    optimizer.zero_grad()       # Zero gradients
    loss = ode_loss(model, t_data, k, N0)  # Compute loss
    loss.backward()             # Backpropagation
    optimizer.step()            # Update parameters
    
    if epoch % 10000 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item()}")

```

```python
# Compare with analytical solution
t_test = torch.linspace(0, 200, 50000).view(-1, 1)
N_pred = model.forward(t_test).detach().numpy()
N_true = N0 * np.exp(-k * t_test.numpy())  # Analytical solution

# Plot results
plt.plot(t_test.numpy(), N_pred, label="Predicted Solution")
plt.plot(t_test.numpy(), N_true, label="Analytical Solution", linestyle="--")
plt.xlabel("t")
plt.ylabel("N(t)")
plt.legend()
plt.title("Solution of ODE using PyTorch Autograd (Polynomial Parameterization)")
plt.show()
```

```python
# Compute the half-life
t_half = np.log(2) / k  # Using the analytical formula for half-life

# Evaluate the model at t_half
N_at_half_life = model.forward(torch.tensor([[t_half]], requires_grad=True)).detach().numpy()

# Print the result
print(f"Half-life (t_1/2) (Actual): {t_half:.2f} seconds")
print(f"Number of atoms left at half-life (predicted): {N_at_half_life[0][0]:.4f}")
```
